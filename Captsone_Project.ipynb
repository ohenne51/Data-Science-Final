```python
import pandas as pd
survey_df = pd.read_csv('/content/drive/MyDrive/Homework/survey_results_public.csv')

survey_df
```

```python
survey_df = survey_df[survey_df['Country'] == 'United States of America']
display(survey_df)
```

```python
# Delete rows where 'ConvertedCompYearly' is blank
survey_df = survey_df.dropna(subset=['ConvertedCompYearly'])
survey_df
```

```python
# Keep only Employment = 'Employed'
survey_df = survey_df[survey_df['Employment'] == 'Employed']
survey_df
```

```python
# Keep salaries between 30,000 and 500,000
survey_df = survey_df[(survey_df['ConvertedCompYearly'] >= 30000) & (survey_df['ConvertedCompYearly'] <= 500000)]
survey_df
```

```python
# Keep YearsCodePro between 1 and 30 years
survey_df = survey_df[(survey_df['YearsCode'] >= 1) & (survey_df['YearsCode'] <= 30)]
survey_df
```

```python
# Save the cleaned sheet as a csv
import pandas as pd
from google.colab import files
survey_df.to_csv('survey_USA_cleaned.csv', index=False)
files.download('survey_USA_cleaned.csv')
```

```python
# Identify top 6 most common industries
industry_counts = survey_df['Industry'].value_counts().head(6)
industry_counts
```

```python
# Display the dataset with only the top 6 industries and only the variables Industry, ConvertedCompYearly, and YearsCode
survey_df = survey_df[['Industry', 'ConvertedCompYearly', 'YearsCode']]
survey_df = survey_df[survey_df['Industry'].isin(industry_counts.index)]
survey_df
```

```python
# Control for experience of 5-10 years
survey_df = survey_df[(survey_df['YearsCode'] >= 5) & (survey_df['YearsCode'] <= 10)]
survey_df
```

```python
# Calculate the median salary by industry
median_salary_by_industry = survey_df.groupby('Industry')['ConvertedCompYearly'].median()
median_salary_by_industry
```

```python
# Display describe descriptive statistics for each industry
industry_stats = survey_df.groupby('Industry')['ConvertedCompYearly'].describe()
industry_stats
```

```python
import matplotlib.pyplot as plt
```

```python
# Make a horizontal bar chart of median salaries by industry
plt.barh(industry_stats.index, industry_stats['50%'])
plt.xlabel('Median Salary')
plt.ylabel('Industry')
plt.title('Median Salaries by Industry')
plt.show()
```

```python
# Make a box plot to illustrate salary distribution by industry
plt.boxplot(survey_df.groupby('Industry')['ConvertedCompYearly'].apply(list))
plt.xlabel('Industry')
plt.ylabel('Salary')
plt.title('Salary Distribution by Industry')
```

```python
# Conduct a one-way ANOVA test for significant differences
from scipy.stats import f_oneway
industry_groups = [survey_df[survey_df['Industry'] == industry]['ConvertedCompYearly'] for industry in industry_stats.index]
f_statistic, p_value = f_oneway(*industry_groups)
print(f'F-statistic: {f_statistic:.2f}')
print(f'p-value: {p_value:.4f}')
```
